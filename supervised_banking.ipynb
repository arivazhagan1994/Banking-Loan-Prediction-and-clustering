# Loan Risk Prediction Project Documentation

## Project Overview
This project aims to build a machine learning model to predict loan risk based on various applicant features. By analyzing historical loan data, the model will classify loan applications as either 'High Risk' or 'Low Risk', helping financial institutions make informed lending decisions and mitigate potential losses.

## Problem Statement
The primary problem addressed is the need for an automated and accurate system to assess the risk associated with loan applications. Manual risk assessment can be time-consuming, inconsistent, and prone to human error. A reliable predictive model is crucial for efficient and effective risk management in the lending process.

## Objectives
The main objectives of this project are:
- To explore and understand the loan application dataset.
- To preprocess the data by handling missing values, encoding categorical variables, and scaling numerical features.
- To perform exploratory data analysis to identify patterns, relationships, and insights within the data.
- To engineer relevant features that can improve model performance.
- To build and evaluate various machine learning models for loan risk prediction.
- To select the best-performing model based on appropriate evaluation metrics.
- To tune and optimize the selected model's hyperparameters for better performance.
- To understand the model's predictions through explainability techniques.
- To validate the model's robustness and generalization ability.
- To outline a deployment strategy for the final model.
- To identify the limitations of the current model and suggest future enhancements.

## Dataset Description
The dataset contains information about loan applicants and includes features such as Income, Age, Experience, Marital Status, House Ownership, Car Ownership, Profession, City, State, Current Job Years, Current House Years, and the target variable, Risk Flag (0 for Low Risk, 1 for High Risk). The dataset has 252000 rows and 13 columns.

## Data Preprocessing
Data preprocessing is a crucial step to prepare the raw data for machine learning models. This involves several sub-steps:

### Handling Missing Values
The dataset was checked for missing values using `df.isnull().sum()`. Fortunately, no missing values were found in this dataset, so no imputation or removal of missing data was required.

### Encoding Categorical Variables
Categorical variables were converted into numerical format using appropriate encoding techniques:
- **Label Encoding:** Applied to ordinal categorical variables ('Married/Single' and 'Age_Group') where there is an inherent order or simply two categories.
- **One-Hot Encoding:** Applied to nominal categorical variables ('House_Ownership', 'Car_Ownership', 'Zone', and 'Profession_Category') where there is no inherent order between the categories. This creates new binary columns for each unique category.

### Feature Scaling & Normalization
Numerical features ('Income', 'Experience', 'CURRENT_JOB_YRS', 'CURRENT_HOUSE_YRS') were scaled using `StandardScaler`. This process standardizes the features to have a mean of 0 and a standard deviation of 1, ensuring that no single feature dominates the model training due to its scale.

## Exploratory Data Analysis (EDA)
EDA was performed to gain insights into the dataset's characteristics and relationships between variables.

### Distribution Analysis
Histograms and box plots were used to visualize the distribution of numerical features and identify potential outliers.

### Correlation Analysis
A heatmap of the correlation matrix was generated to visualize the relationships between numerical features and the target variable. This helps in understanding which features are most strongly correlated with loan risk.

## Feature Engineering
New features were created from existing ones to potentially improve model performance.

### Age Grouping
The 'Age' column was grouped into different age bins ('20-25', '26-30', etc.) to create a new categorical feature 'Age_Group'. This can help capture non-linear relationships between age and risk.

## Model Selection
Several machine learning models were considered for the loan risk prediction task:

### Logistic Regression
A linear model that estimates the probability of the target variable belonging to a particular class. It's a good baseline model for binary classification.

### Random Forest
An ensemble learning method that constructs multiple decision trees during training and outputs the mode of the classes (classification) or mean prediction (regression) of the individual trees. It's known for its robustness and ability to handle non-linear relationships.

### XGBoost
Extreme Gradient Boosting is a powerful and efficient implementation of gradient boosting. It's widely used for its high performance and ability to handle various data types and complexities.

## Model Comparison
The performance of the selected models was compared based on various evaluation metrics.

## Model Evaluation Metrics
The following metrics were used to evaluate the performance of the classification models:

### Accuracy
The proportion of correctly predicted instances out of the total number of instances. While useful, it can be misleading in imbalanced datasets.

### Precision
The ratio of true positives to the total predicted positives. It measures the accuracy of the positive predictions.

### Recall
The ratio of true positives to the total actual positives. It measures the ability of the model to find all the positive instances.

### F1-Score
The harmonic mean of precision and recall. It provides a balanced measure of the model's performance, especially in the presence of class imbalance.

### ROC-AUC
Receiver Operating Characteristic - Area Under the Curve. It measures the ability of the model to distinguish between the positive and negative classes. A higher AUC indicates better discriminatory power.

## Hyperparameter Tuning & Optimization
`GridSearchCV` cross-validation was used to tune the hyperparameters of the selected model (XGBoost). This process involves searching over a predefined grid of hyperparameter values to find the combination that yields the best performance (based on the chosen scoring metric, e.g., F1-score) on the validation sets.

## Model Explainability
Understanding why a model makes certain predictions is crucial, especially in sensitive domains like finance.

## Model Validation
The model's performance was validated using appropriate techniques.

### Cross-Validation
Stratified K-Fold cross-validation was used during hyperparameter tuning to ensure the model's performance is consistent across different subsets of the data and to get a more reliable estimate of its generalization ability.

### Train-Test Split Results
The final model was evaluated on a separate test set that was not used during training or tuning. The performance metrics (Accuracy, Precision, Recall, F1-Score, ROC-AUC) on the test set provide an unbiased evaluation of the model's performance on unseen data.

## Final Model Selection
Based on the evaluation metrics and the goals of the project (e.g., minimizing false negatives for high-risk predictions), the XGBoost model with optimized hyperparameters was selected as the final model.

## Deployment Strategy
A strategy for deploying the trained model for making predictions on new loan applications was outlined.

### Saving Model with Pickle
The trained model (`best_model`) and the scaler (`scaler`) were saved using `joblib` (or `pickle`) to allow for easy loading and use in a production environment without retraining.

### Using StandardScaler
The saved `StandardScaler` object will be used to scale the numerical features of new incoming data before making predictions, ensuring consistency with the data the model was trained on.

### Categorical Inputs (One-Hot / Label Encoding)
New categorical inputs will need to be encoded using the same `LabelEncoder` and one-hot encoding steps applied during training to ensure the input format matches the model's expectations.

### Risk Flag (High Risk / Low Risk)
The model's output (0 or 1) will be translated into a clear 'Low Risk' or 'High Risk' classification for easy interpretation.

### Probability Score
The model can also output a probability score (using `predict_proba`) which can provide more nuanced information about the level of risk. This probability can be used to set thresholds for classification or provide additional context to decision-makers.

### Visualization in Deployment
Visualizations can be incorporated into the deployment to help users understand the model's output.

#### Probability Chart
A chart showing the probability of being high risk can be displayed.

#### Feature Impact Visualization
Visualizations showing the impact of individual features on the prediction for a specific applicant (e.g., using SHAP values) can provide transparency and build trust in the model.

The outlined deployment strategy and considerations for model explainability and visualization aim to facilitate the practical application of the model in a real-world setting.
```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[column] = df[column].apply(lambda x:lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)
    return df
```

```python
# Separate the 'STATE' column into Zone wise categories
# Define the mapping of states to zones

zone_mapping = {
    'North': [
        'Delhi', 'Punjab', 'Haryana', 'Uttarakhand', 'Himachal_Pradesh',
        'Jammu_and_Kashmir', 'Chandigarh', 'Uttar_Pradesh', 'Uttar_Pradesh[5]'
    ],
    'South': [
        'Tamil_Nadu', 'Karnataka', 'Andhra_Pradesh', 'Telangana',
        'Kerala', 'Puducherry'
    ],
    'East': [
        'West_Bengal', 'Bihar', 'Odisha', 'Jharkhand', 'Assam', 'Tripura',
        'Manipur', 'Sikkim', 'Mizoram'
    ],
    'West': [
        'Maharashtra', 'Gujarat', 'Rajasthan', 'Goa', 'Madhya_Pradesh', 'Chhattisgarh'
    ]
}

# Create a new column 'Zone' based on the mapping
def map_zone(state):
    for zone, states in zone_mapping.items():
        if state in states:
            return zone
    return 'Other'  # For any state not in the mapping, assign 'Other' zone

df['Zone'] = df['STATE'].apply(map_zone)
df.head()
```

How to Choose label encoding or one hot encoding
- One-Hot Encoding when categories have no order (e.g., City, Profession).
- Label Encoding when categories have a clear order (e.g., Low < Medium < High).

```python
df1.info()
```

```python
#Label encoding for categorical columns
# what i the label encoding
'''Label encoding is a technique used to convert categorical variables into numerical format.
Each unique category is assigned a unique integer value,
which allows machine learning algorithms to process the data more effectively.
It is particularly useful for ordinal categorical variables where the order matters.
Example: "Married" might be encoded as 1 and "Single" as 0, allowing algorithms to interpret the relationship between these categories.
'''
le = LabelEncoder()
df1['Married/Single'] = le.fit_transform(df1['Married/Single'])
df1['Age_Group'] = le.fit_transform(df1['Age_Group'])
```

```python
# One-Hot Encoding when categories have no order (e.g., City, Profession).
df1 = pd.get_dummies(df1, columns=['House_Ownership', 'Car_Ownership', 'Zone', 'Profession_Category'], drop_first=True)
df1.info()
```

Scaling and Normalization

```python
# Scaling and Normalization
# Scaling is the process of transforming features to a similar scale, which helps improve the performance of many machine learning algorithms.
# Normalization is a specific type of scaling that adjusts the values in a dataset to a common scale, typically between 0 and 1.
# It is particularly useful when features have different units or scales.
# Example: Scaling income and age to ensure they contribute equally to model training.

numeric_col = ['Income', 'Experience', 'CURRENT_JOB_YRS', 'CURRENT_HOUSE_YRS']

# Plot before scaling
plt.figure(figsize=(16, 6))
df1[numeric_col].hist(bins=30,figsize=(16, 6),layout=(2, 3))
plt.suptitle('Before Scaling')

scaler = StandardScaler()
df_scaled = df1.copy()
df_scaled[numeric_col] = scaler.fit_transform(df1[numeric_col])

#  Plot after scaling
plt.figure(figsize=(12, 6))
df_scaled[numeric_col].hist(bins=30, figsize=(12, 6), layout=(2, 3))
plt.suptitle(" After Scaling (StandardScaler)", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
xgb = XGBClassifier(
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42,
)

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 10],
    'learning_rate': [0.01, 0.1, 0.2],
    'scale_pos_weight': [1, 2, 3]
}
```

```python
grid = GridSearchCV(
    estimator = xgb,
    param_grid = param_grid,
    scoring = 'f1',
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    verbose = 1,
    n_jobs = -1
)

grid.fit(X_train, y_train)
print("Best parameters found: ", grid.best_params_)
print("Best cross-validation score: ", grid.best_score_)

best_model = grid.best_estimator_
y_pred = best_model.predict(X_test)

acc = accuracy_score(y_test, y_pred)
pric = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Model Accuracy: {acc:.2f}")
print(f"Precision: {pric:.2f}")
print(f"Recall: {rec:.2f}")
print(f"F1 Score: {f1:.2f}")
```
